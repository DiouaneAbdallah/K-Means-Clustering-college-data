
# K Means avec  Python


## La méthode utilisée 
K-means est une méthode d'apprentissage insupervisé qui essaie de partitionner les données en se basant sur la similarité(distance). L'apprentissage insupervisé veut dire qu'il n'y a pas de sortie à prédire et que l'algorithme essaie de dévouvrir les structures existantes dans les données. Dans K-means, nous devons spécifier le nombre de clusters. L'algorithme assigne aléatoirement les centres des clusters. Ensuite, pour chaque itération qui suit: il affecte chaque donnée au centre le plus proche, il calcule le nouveau centre de cluster et ainsi de suite. L'algorithme continue tant que les centres des clusters sont en train de changer de position.



## Importation de librairies

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creation de données

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
data = make_blobs(n_samples=1000, n_features=4, 
                           centers=6, cluster_std=1.8,random_state=101)
```

## Visualisation de données 

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

## Création de clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
plt.show()
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
#ax1.scatter(*zip(*kmeans.cluster_centers_), color='black', marker='x', linewidth=3)

ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

La différence de couleurs entre les données des deux figures n'a pas de signification
